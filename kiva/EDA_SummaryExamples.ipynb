deskriptive Part

In dem Piechart ist die Verteilung unserer Kundengruppen abgebildet. Je größer ein Wedge desto größer der Kundenanteil.

In dem Piechart ist zu sehen, dass unser Unternehmen zwei Kundengruppen hat. 15% machen die Customer aus und 85% die Subscriber. Hierbei wurden rund 670.000 Daten ausgewertet.

Hier ist die Frage interessant, wer hinter diesen Kundengruppen steckt!

analytische Part

Sollzustand

Um den Sollzustand zu ermitteln, kann das Bezahlmodell hinzugezogen werden. Wenn Subscriber einen festen Betrag zahlen und die Räder so oft nutzen können, wie sie wollen, dann sorgt der überweigende Anteil der Dauermieter für einen regelmäßigen Cashflow, womit wiederum die Fixkosten bezahlt werden. Allerdings kann nicht gesagt werden, wie viele Subscriber hinter den 85% stecken. Im Extremfall könnte auch 1 Subscriber dahinter stekcen, der sich ständig Räder ausleiht.

Wäre der Anteil der Customer größer und zahlen diese nur für die Zeit in der sie die Räder nutzen, dann haben wir ein volatiles Geschäftsmodell.

Vergleich Soll- und Istzustand

85% meiner Kunden bestehen aus Subscribern. Unter den bisher getroffenen Annahmen ist dies erst einmal positiv, da das Unternehmen regelmäßig Geld bekommt und kein volatiles Geschäftsmodell vorliegt.

Was die Daten hier nicht zeigen, ist die Zusammensetzung der Kunden, also wie viele Subscriber wir haben. Es ist nur die Anzahl der Trips zu sehen, aber nicht wie oft ein Subscriber x sich tatsächlich ein Rad ausgeliehen hat. Hier könnte eine Kunden_id hilfreich sein. Ob diese aus datenschutzgründen zur Verfügung gestellt wird, bleibt fraglich.

Auch wenn wir regelmäßig Geld bekommen, kommt die Frage auf, wann genau die Räder genutzt werden. Es bringt nichts, wenn wir zwar Geld bekommen, aber unsere Fahrräder nur am Samstag oder Sonntag genutzt werden. Vergleich Soll- und Istzustand

In der Grafik ist zu sehen, dass die blauen Balken von Montag bis Freitag eine gute Nutzung haben. Dass bedeutet, dass die Subscriber unsere Fahrräder regelmäßig unter der Woche nutzen.
