```python
def count_male_occurrences(text):
    if not isinstance(text, str):
        return;
    male_pattern = re.compile(r'\bmale\b', flags=re.IGNORECASE)
    male_count = len(male_pattern.findall(text))
    return male_count

def count_female_occurrences(text):
    if not isinstance(text, str):
        return;
    female_pattern = re.compile(r'\bfemale\b', flags=re.IGNORECASE)
    female_count = len(female_pattern.findall(text))
    return female_count

df['male_count']=df['borrower_genders'].map( count_male_occurrences )
df['female_count']=df['borrower_genders'].map( count_female_occurrences )
bucketsize = 5
df['borrower_count']= (( df['male_count'] + df['female_count'] ) // bucketsize)*bucketsize
map = {0: "0-5",5:"5-10", 10:"10-15", 15:"15-20", 20:"20-25", 25:"25-30"} 
df['borrower_count'] = df['borrower_count'].map( map )

categoryarrayAll = ["Agriculture", "Food","Arts","Retail","Education","Construction","Construction", 
                    "Manufacturing","Clothing","Entertainment","Transportation","Personal Use", "Services","Health","Housing", "Wholesale"]

mostPopularCategories = ["Agriculture", "Food","Arts","Retail","Education", 
                            "Clothing","Transportation","Personal Use", "Services","Housing"]

df_selected = df.loc[ (df["country"]=="Peru") & (df["sector"].isin(mostPopularCategories)) , : ] # df["borrower_count"] <= 30) &

df_grouped = df_selected.groupby(by=["sector","borrower_count"],as_index=False ).agg( sum_loan_amount=("loan_amount","median") )
df_grouped

df_matrix = df_grouped.pivot_table(index='borrower_count', columns='sector', values='sum_loan_amount', aggfunc='median')
df_matrix = df_matrix.sort_index( ascending=False)

df_matrix_sorted  = df_matrix.loc[:,mostPopularCategories]

fig = px.imshow(df_matrix_sorted, 
                
                color_continuous_scale="YlOrRd", 
                labels={"y": "Anzahl der am Projekt Beteiligten", "x":"Oberkategory der Projekte"},
                title="Die Höhe der Zielbeträge korrelieren mit der Anzahl der Projektbeteiligten", 
                text_auto= True,)

fig.update_yaxes(autorange=True)
# fig.update_xaxes(categoryorder = "total descending", categoryarray=categoryarray)

fig.show()

df_grouped_sector = df.groupby(by="sector", as_index=False).agg( count = ("sector", "count") )
fig = px.bar( df_grouped_sector, x="sector", y="count" )
fig.update_xaxes( categoryorder = "total descending" )
fig.show()


df_matrix_sorted
map.values()
```
